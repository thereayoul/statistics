```python
df_normal = df.resample(rule='D').cnt.sum()
df_normal
```

```python
df_avg = df.resample(rule='D').cnt.sum().rolling(window=3).mean()
df_avg
```

```python
df_new = df_normal - df_avg
```

```python
df_avg_new = df_avg.to_frame(name='cnt').reset_index()
df_avg_new
```

```python
df_avg_new.query('timestamp == "2015-07-09"')
```

```python
df_new_new = df_new.to_frame(name='cnt').reset_index()
df_new_new
```

# 6

```python
std = np.std(df_new_new['cnt'])
std
```

# 7

```python
upper_bound = df_avg_new['cnt'] + 2.576 * std
upper_bound = upper_bound.to_frame(name='cnt')
upper_bound
```

```python
lower_bound = df_avg_new['cnt'] - 2.576 * std
lower_bound = lower_bound.to_frame(name='cnt')
lower_bound
```
